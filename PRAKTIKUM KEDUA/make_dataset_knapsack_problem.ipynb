```python
import pandas as pd
import numpy as np
from numpy.random import default_rng
```

# BUAT DATASET K-PROBLEM
1. DATASET N = 25
2. DATASET N = 50
3. DATASET N = 100
4. DATASET N = 200
5. DATASET N = 400

```python
rng = default_rng()
def bangkitkan_data(total_data):
    return rng.integers(1, 100, size=total_data)

def buat_dataset(total_data):
    w = bangkitkan_data(total_data)
    y = bangkitkan_data(total_data)
    d = pd.DataFrame(columns=['w', 'y'])
    d['w'] = w
    d['y'] = y
    name = 'dataset_knapsack_' + str(total_data) + '.csv'
    d.to_csv(name, index=False)
    return d

d_25 = buat_dataset(25)
d_50 = buat_dataset(50)
d_100 = buat_dataset(100)
d_200 = buat_dataset(200)
d_400 = buat_dataset(400)
```
