```python
import pandas as pd 
import numpy as np
from numpy.random import default_rng
```

# BUAT DATASET ASSIGMENT PROBLEM
1. DATASET NJOB = 10 & NPERSON = 10
2. DATASET NJOB = 10 & NPERSON = 20
3. DATASET NJOB = 20 & NPERSON = 10
4. DATASET NJOB = 20 & NPERSON = 20
5. DATASET NJOB = 40 & NPERSON = 40

```python
rng = default_rng()
def bangkitkan_data(total_data):
    return rng.integers(0, 100, size=total_data)

def buat_dataset(nperson, njob):
    total_data = njob * nperson
    data = bangkitkan_data(total_data)
    data = data.reshape(nperson, njob)
    data = pd.DataFrame(data)
    data.columns = [f'J{i}' for i in range(1, njob+1)]
    data.to_csv(f'dataset_{nperson}_{njob}.csv', index=False)
    return data

d_10_10 = buat_dataset(10, 10)
d_10_20 = buat_dataset(10, 20)
d_20_10 = buat_dataset(20, 10)
d_20_20 = buat_dataset(20, 20)
d_40_40 = buat_dataset(40, 40)
    
```
